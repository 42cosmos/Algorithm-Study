```python
df = pd.read_excel(test_list_filename, header = 2).iloc[:-2, :-1]
df
```

```python
def scraping_desc(url:str):    
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    
    web = '/Users/eunbinpark/workspace/chromedriver'
    
    driver = webdriver.Chrome(web, options=options)
    driver.get(url)
    
    desc_list = list()

    try:
        desc_list.append(driver.find_element_by_xpath('//*[@id="tour2"]/div/div/p[1]').text)
        desc_list.append(driver.find_element_by_xpath('//*[@id="tour2"]/div/div/p[2]').text)

    except:
        pass
    
    driver.close()
    
    return desc_list
```

```python
descriptions_list = df['링크'].progress_map(scraping_desc).tolist()
```

```python
len(descriptions_list)
```

```python
def is_exists_dir(dir_path, directory):
    directory = os.path.join(dir_path, directory)
    if not os.path.exists(directory):
        os.makedirs(directory)
        print('Create Dir' + directory)
            
    else:
        print('Already Exists' + directory)

        
def make_prob2md_file(day_count, prob_title, description, prob_url):
    md_format = f"""## Day {day_count}\n
### {prob_title}
{description}<br><br>
### URL
[{prob_title}]({prob_url})<br>"""
    
    return md_format        
```

```python
dir_names = ('Day ' + df['날짜'].map(lambda x: x.split()[0])).tolist()

for dir_name in dir_names:
    is_exists_dir(path_for_readme, dir_name)
```

```python
day_counts = df['날짜'].map(lambda x: x.split()[0]).tolist()
prob_titles = df['문제이름'].tolist()
# descriptions_list
urls = df['링크'].tolist()
```

```python
final_df = pd.DataFrame(data = [dir_names, day_counts, prob_titles, descriptions_list, urls], index = ['dir_names', 'day_counts', 'prob_titles', 'desc_list', 'urls']).T
final_df['desc_list'] = final_df['desc_list'].map(lambda x: '<br>'.join(x))
final_df
```

```python
for i in final_df.itertuples():
    dir_name = i[1]
    day_count = i[2]
    prob_title = i[3]
    description = i[4]
    url = i[5]
    
    with open(os.path.join(path_for_readme, dir_name, 'README.md'), 'w') as f:
        f.write(make_prob2md_file(day_count, prob_title, description, url))
```
